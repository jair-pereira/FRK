```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
import pandas as pd
%matplotlib inline

rcParams['figure.dpi'] = 100
```

```python
##PLOT GROUPED BY DIMENSION
experiment_set = ["jpnsec_0101", "jpnsec_0201",
                  "jpnsec_1501", "jpnsec_1701"]

dim = ["02", "05"]
colors = ["red", "blue"]
label = ["D2", "D5"]
titles = ["Function 01", "Function 02", "Function 15", "Function 17"]

for j, experiment_name in enumerate(experiment_set):
    ## plot settings ##
    fig, ax = plt.subplots()
    title = "Minimum fitness"
    xlabel = "Generations"
    ylabel = "Fitness"
    x = [i for i in range(0,31)]
    
    for k in range(2):
        ## read files ##
        base = pd.read_csv("./"+experiment_name+dim[k]+"_10/history.csv")[["gen", "hh_fit"]]#.groupby('gen')
        for i in range(1, 10):
            base = base.append(pd.read_csv("./"+experiment_name+dim[k]+"_"+str(i)+"/history.csv")[["gen", "hh_fit"]])

        ## plot ##
        y = base.groupby('gen').min()["hh_fit"]
        ax.plot(x, y, lw=2, color=colors[k], label=label[k])
    
    ax.set(xlabel=xlabel, ylabel=ylabel, title=titles[j]+" instance 1")
    plt.legend()
    ax.grid()
    plt.show()
```
